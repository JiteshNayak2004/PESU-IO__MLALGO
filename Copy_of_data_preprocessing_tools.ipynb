# PREDICTING STUDENT ADMISSION 

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_squared_error
from sklearn import metrics
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor 
from sklearn.ensemble import RandomForestRegressor
```

## Importing the dataset

```
dataset = pd.read_csv('MLPROJ.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
gre=dataset.iloc[:,1].values
toefl=dataset.iloc[:,2].values
```

```
print(X)
```

```
print(y)
```

SPLITTING DATASET

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

MULTI LINEAR REGRESSION

```
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)

#code for customized input
a=regressor.predict([[337,118,4,4.5,4,9.2]])
print(a)
```

```
arr=[]
for i in range(0,80):
  arr.append(i)
print(arr)
plt.plot(arr,y_test)
plt.xlabel('no of elements')
plt.ylabel('test values and predicted values')
plt.plot(arr,y_pred)
plt.title('Test values in Blue and Predicted values in orange')



```

INFERENCE

```
plt.scatter(x=y_test, y=y_pred);
plt.xlabel('test values')
plt.ylabel('predicted values')
print("the mean squared error is",mean_squared_error(y_test, y_pred))
print('\n')
err_mll=mean_squared_error(y_test, y_pred)
```

KNN REGRESSION

```
from sklearn.neighbors import KNeighborsRegressor
KNneigh = KNeighborsRegressor(n_neighbors=2)
KNneigh.fit(X_train,y_train)
knn= KNneigh.predict(X_test)
```

VISUALIZING THE KNN MODEL

```
plt.plot(arr,y_test)
plt.xlabel('no of elements')
plt.ylabel('test values and predicted values')
plt.plot(arr,knn)
plt.title('Test values in Blue and Predicted values in orange')
```

INFERENCE

```
plt.scatter(x=y_test, y=knn);
plt.xlabel('test values')
plt.ylabel('predicted values')
print("the mean squared error is",mean_squared_error(y_test, knn))
print('\n')
err_knn=mean_squared_error(y_test, knn)
```

RANDOM FOREST REGRESSION

```

RF = RandomForestRegressor(n_estimators = 100)
RF.fit(X_train,y_train) 
randfor= RF.predict(X_test)
```

VISUALISING THE RANDOM FOREST MODEL

```
plt.plot(arr,y_test)
plt.xlabel('no of elements')
plt.ylabel('test values and predicted values')
plt.plot(arr,randfor)
plt.title('Test values in Blue and Predicted values in orange')
```

INFERENCE

```
plt.scatter(x=y_test, y=randfor);
plt.xlabel('test values')
plt.ylabel('predicted values')
print("the mean squared error is",mean_squared_error(y_test, randfor))
print('\n')
err_randfor=mean_squared_error(y_test, randfor)
```

DECISION TREE REGRESSION

```

DTR = DecisionTreeRegressor() 
DTR.fit(X_train,y_train)
dectree= DTR.predict(X_test)
```

VISUALISING DATA TREE REGRESSION

```
plt.plot(arr,y_test)
plt.xlabel('no of elements')
plt.ylabel('test values and predicted values')
plt.plot(arr,dectree)
plt.title('Test values in Blue and Predicted values in orange')
```

INFERENCE

```
plt.scatter(x=y_test, y=dectree);
plt.xlabel('test values')
plt.ylabel('predicted values')
print("the mean squared error is",mean_squared_error(y_test, dectree))
print('\n')
err_dectree=mean_squared_error(y_test, dectree)
```

FINAL INFERENCES

```
print("The best model that fits our dataset is multi linear regression with \na mean squared error of",min(err_mll,err_randfor,err_knn,err_dectree))
```
